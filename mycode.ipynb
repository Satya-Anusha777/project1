```python
new_df = pd.get_dummies(df,columns = ['gender'])
new_df = df.iloc[:,2:6]
new_df = pd.get_dummies(new_df,columns = ['gender'])
new_df
```

```python
y = df["grade"]
x = new_df
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_training_set ,X_testing_set,Y_training_set,Y_testing_set = train_test_split(x,y,test_size = 25,random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_training_set,Y_training_set)
regressor.score(X_training_set,Y_training_set)
print('Coefficients:\n',regressor.coef_)
```

```python
y_pred = regressor.predict(X_testing_set)
y_test2 = pd.DataFrame(Y_testing_set)
y_pred2 = pd.DataFrame(y_pred)
y_test2.index = y_pred2.index
result = pd.concat((y_test2,y_pred2),join = 'outer',axis = 1)
Result = pd.concat([y_test2,y_pred2],axis = 1,join_axes = [y_test2.index])
print(Result)
```
